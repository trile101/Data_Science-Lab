### 1. Phân tích đơn biến

##### Dựa vào thuôc tính ```TotalCases``` và ```Date``` qua đó ta sẽ phân tích số ca nhiễm trên thế giới diễn ra trong 1 tuần

````python
days = data["Date"].unique()

fig, ax = plt.subplots(figsize=(10, 5))
Sum_day = []

for day in days:
    temp = data.loc[data["Date"]==day]
    Sum_day.append(temp["TotalCases"].sum())

plt.title("Số ca nhiễm covid-19 của thế giới trong 7 ngày (29/11/2021 - 05/12/2021)")
plt.plot(days, Sum_day, color='red', marker='o')
plt.xlabel('Ngày')
plt.ylabel('Ca nhiễm')
plt.grid(True)
plt.show()
````

- Dựa vào biểu đồ đường trên ta thấy được xu hướng của dữ liệu. Ở đây, ta thấy được số ca nhiễm trong 1 tuần tăng khá nhanh, số lượng ca nhiễm mới trong ngày 03/12 và 04/12 dường như tăng gấp 3 lần so với các ngày trước

##### Phân tích các ca nhiễm mới
- Ta sẽ dựa vào biểu đồ hình hộp để biểu diễn tả 5 vị trí phân bố dữ liệu (giá trị nhỏ nhất (min), tứ phân vị thứ nhất (Q1), trung vị (median), tứ phân vị thứ 3 (Q3) và giá trị lớn nhất (max)) để hiển thị sư phân bố của ca nhiễm mới của các nước trên thế giới qua từng ngày (cụ thể từ 30/11/2021 - 05/12/2021) . Đồng thời ta cũng có thể xem thử lý do tại sao số ca nhiễm mới lại tăng đột ngột trong ngày 3/12 và 4/12

````python
days = data["Date"].unique()

temp = data.loc[data["Date"]==days[0]]
value_box = []
for day in days[1:]:
    temp1 = data.loc[data["Date"]==day]
    value_box.append(temp1["TotalCases"].to_numpy() - temp["TotalCases"].to_numpy())
    temp = temp1


plt.figure(figsize=(10, 10))
plt.boxplot(value_box)
plt.xticks([1, 2, 3, 4, 5, 6], days[1:])

plt.title("Biểu đồ phân phối ca nhiễm mới từ ngày 29/11/2021 - 05/12/2021")
plt.ylabel("Ca nhiễm mới")
plt.xlabel("Ngày")
plt.show()
````

- Dựa vào biểu đồ hộp ta thấy được phân bố ca nhiễm mới của các nước qua từng ngày, trong 6 ngày số ca nhiễm mới của các nước thuộc dạng outlier rất nhiều, đặc biệt ta thấy trong ngày 3/12 và 4/12 có 1 vài outlier so với trung bình, nó có khảng cách rất lớn , nên ta có thể suy luận rằng: ```Có 1 vài nước có số ca nhiễm tăng đột ngột dẫn đến tổng số ca nhiễm trong ngày đó tăng mạnh hơn so với các ngày trước```

#### Phân tích phần trăm của 6 châu lục có số ca nhiễm covid-19 trong ngày  03/12/2021

````python
day = "2021-12-03"
temp_data = data.loc[(data["Date"]==day)]
continents = data["Continent"].unique()

print("Số lượng nước trên mỗi châu lục")
for continent in continents:
    t = temp_data.loc[temp_data["Continent"]==continent]
    print("-", continent, ": ", t.shape[0], " nước")
````

- Vẽ biểu đồ tròn ca nhiễm covid-19 của 6 châu lục 

````python
per_continent = []
continents = data["Continent"].unique()
Sum = temp_data["TotalCases"].sum()

for continent in continents[0:5]:
    c = temp_data.loc[data["Continent"]==continent]
    per_continent.append(round(c["TotalCases"].sum()/ Sum, 2))

per_continent.append(round(1 - sum(per_continent), 2))

fig, ax = plt.subplots(figsize =(10, 7))

ax.pie(per_continent, labels = continents, autopct='%1.2f%%')
ax.legend(title = "Six continents", loc ="best", bbox_to_anchor =(1, 0, 0.3, 1))
ax.set_title('Biểu đồ tròn về phần trăm ca nhiễm covid-19 của 6 châu lục')
plt.show()
````

- Dựa vào biểu đồ tròn ta thấy rõ ràng rằng 4 châu lục (châu á, châu âu, bắc mĩ và nam mĩ) có lương phần trăm lớn

#### Phân tích tỉ lệ tử vong của 4 châu lục (châu á, châu âu, bắc mĩ và nam mĩ) trong 1 tuần
- Do biểu đồ tròn, ta thấy 4 châu lục trên có tỉ lệ nhiễm covid-19 cao, ta sẽ tiến hành phân tích 4 châu lục đó

````python
continents = ["North America", "Asia", "South America", "Europe"]

days = data["Date"].unique()

plt.figure(figsize=(13, 7))

for continent in continents: 
    # cal date
    value = []
    for day in days:
        t = data.loc[(data['Continent']==continent) & (data['Date']==day)]
        value.append(t["TotalDeaths"].sum())

    plt.plot(days, value, label=continent)
    plt.title("Tổng số ca tử vong covid-19 trong 1 tuần (29/11/2021 - 05/11/2021)")

plt.legend(loc ="best")
    
plt.show()
````

- Ta thấy được số ca tử vong của châu âu lớn hơn nhiều so với 3 châu còn lại. Đồng thời số lượng ca tử vong của châu âu tăng mạnh trong 1 tuần và số lượng ca tử vong ở 3 châu còn lại tăng không nhiều

### 2. Phân tích đa biến

#### Ta tiến hành sử dụng biểu đồ nhiệt (Heat Map) để thể hiện hệ số tương quan (Correlation coefficient) để đo mối quan hệ giữa 2  thuộc tính bất kì trong 13 thuộc tính
  - Trích từ ```https://vietnambiz.vn```: Sức mạnh của mối quan hệ dựa trên giá trị của hệ số tương quan. Ví dụ: Giá trị 0,2 cho thấy có mối tương quan đồng biến, nhưng nó yếu và không đáng kể. Các chuyên gia cho rằng tương quan có ý nghĩa khi ít nhất đạt giá trị 0,8. Tuy nhiên, hệ số tương quan với giá trị tuyệt đối là 0,9 hoặc lớn hơn sẽ thể hiện mối quan hệ rất mạnh mẽ.

````python
temp = data.loc[data["Date"]=="2021-12-05"]
d = temp.corr(method='pearson')
f, ax = plt.subplots(figsize=(30, 15))
sns.heatmap(d, ax=ax, cmap="YlGnBu", annot=True)
````

- Dựa vào đồ thị có 5 cặp thuộc tính có mối quan hệ rất mạnh mẽ, có hệ số tương quan lớn hơn 0.9
   + TotalCases và TotalDeaths
   + TotalCases và TotalRecovered
   + TotalCases và TotalTest
   + TotalDeaths và TotalRecovered
   + TotalRecoved và TotalTest

#### Ta sử dụng 2 thuộc tính ```TotalCases``` và ```TotalDeaths ``` xét mối quan hệ giữa chúng bằng biểu đồ phân tán (scatter plot)

````python
temp_data = data.loc[data["Date"]=="2021-12-05"]

figure, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 8))


figure.suptitle('Biểu đồ phân tán giữa 2 thuộc tính')
# Add title
ax1.scatter(temp_data['TotalTests'], temp_data['TotalCases'], c ="blue")
ax1.set_xlabel("TotalTests")
ax1.set_ylabel("TotalCases")

ax2.scatter(temp_data['TotalDeaths'],temp_data['TotalRecovered'], c ="blue")
ax2.set_xlabel("TotalDeaths")
ax2.set_ylabel("TotalRecovered")

plt.show()
````

- Qua đồ thị ta thấy rõ ràng 2 thuộc tính có ```mối quan hệ nhân quả``` hay ```mối quan hệ đồng biến```
  + Ở cặp thuộc tính ```TotalTests``` và ```TotalCases```, khi số lượng ca nhiễm tăng nó sẽ dẫn đến số lượng ca tử vong cũng tăng theo, các điểm trong đồ thị hội tụ tuyến tính.
  + Ở cặp thuộc tính ```TotalDeaths``` và ```TotalRecovered```, khi số lượng ca nhiễm bình phục tăng nó sẽ dẫn đến số lượng ca tử vong cũng tăng theo, ta thấy có sự đối ngược giữa lí thuyết và thực tế. 

#### Ta sẽ sử dụng 3 thuộc tính ```Deaths/1M pop``` , ```TotCases/1M pop``` và ```Tests/1M pop``` của các nước trên thế giới cho biểu đồ bong bóng( bubble plot), tiến hành xem thử 3 thuộc tính có mối quan hệ hay không, ta sẽ sử dụng ngày "05/12/2021" để trực quan hóa 

````python
temp_data = data.loc[data["Date"]=="2021-12-05"]
plt.figure(dpi=120)
plt.scatter(temp_data["Deaths/1M pop"], temp_data["TotCases/1M pop"], s=temp_data["Tests/1M pop"]/50000 , alpha=0.5)
plt.show()
````
